```python
#5.Write a program that accepts a sentence and calculate the number of letters and digits.

a = input("enter string : ")
c = 0
s = 0

for i in a:
    if i.isnumeric():
        c = c+1
    elif i.isalpha():
        s = s+1
            
print(f"without space including \ncount of letter is : {s} and count of digit is : {c}")
```
